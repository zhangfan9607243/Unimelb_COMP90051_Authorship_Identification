# Data Processing

## 1. Import Libraries

```python
import math
import numpy as np
import pandas as pd
pd.set_option('future.no_silent_downcasting', True)
```

## 2. Load Original Data

```python
PATH_PREFIX = "../"
```

```python
data_tran = pd.read_json(PATH_PREFIX + "data/data_original/train.json")
data_test = pd.read_json(PATH_PREFIX + "data/data_original/test.json")
```

## 3. Data Preprocessing

### 3.1 Create the 'coauthors' column in training data

```python
data_tran['authors_main'] = data_tran['authors'].apply(lambda x: [a for a in x if 0 <= a < 100])
data_tran['authors_coop'] = data_tran['authors'].apply(lambda x: [a for a in x if a >= 100])
data_tran['authors_main'] = data_tran['authors_main'].apply(lambda x: x if x else [-1])
```

```python
data_test = data_test.rename(columns={'coauthors': 'authors_coop'})
```

### 3.2 Rename list format of text columns and combine them into a new column

```python
data_tran = data_tran.rename(columns={'title': 'list_title', 'abstract': 'list_abstract'})
data_test = data_test.rename(columns={'title': 'list_title', 'abstract': 'list_abstract'})
```

```python
data_tran["list_combined"] = data_tran["list_title"] + data_tran["list_abstract"]
data_test["list_combined"] = data_test["list_title"] + data_test["list_abstract"]
```

### 3.3 Create the string format of title & abstract and combine them into a new column

```python
def list_to_string(lst):
    return ' '.join(map(str, lst))
```

```python
data_tran['str_title'] = data_tran['list_title'].apply(list_to_string)
data_test['str_title'] = data_test['list_title'].apply(list_to_string)

data_tran['str_abstract'] = data_tran['list_abstract'].apply(list_to_string)
data_test['str_abstract'] = data_test['list_abstract'].apply(list_to_string)
```

```python
data_tran['str_combined'] = (data_tran['str_title'] + ' ' + data_tran['str_abstract'])
data_test['str_combined'] = (data_test['str_title'] + ' ' + data_test['str_abstract'])
```

### 3.4 Fill NA in venue

```python
data_tran['venue'] = data_tran['venue'].replace('', 465).fillna(465).astype(int)
data_test['venue'] = data_test['venue'].replace('', 465).fillna(465).astype(int)
```

## 4. Present the Preprocessed Data

### 4.1 Present Training Data

```python
data_tran = data_tran[['authors', 'authors_main', 'authors_coop', 'venue', 'year', 'list_title', 'list_abstract', 'list_combined', 'str_title', 'str_abstract', 'str_combined']]
data_tran.head(5)
```

### 4.2 Present Testing Data

```python
data_test = data_test[['authors_coop', 'venue', 'year', 'list_title', 'list_abstract', 'list_combined', 'str_title', 'str_abstract', 'str_combined']]
data_test.head(5)
```

## 5. Save the Preprocessed Data

### 5.1 Save the training and testing data as json

```python
data_tran.to_json(PATH_PREFIX + 'data/data_processed/data_tran.json', orient='index')
data_test.to_json(PATH_PREFIX + 'data/data_processed/data_test.json', orient='index')
```

### 5.2 Save the label in onehot as npy

```python
def process_label(data):
    
    n_author = 101  
    threshold_author = 100  

    author_list = np.zeros((len(data), n_author))

    for i, authors in enumerate(data["authors"]):
        author_id = np.array(authors)[np.array(authors) < threshold_author]
        
        if len(author_id) == 0:
            author_list[i, -1] = 1
        else:
            author_list[i, author_id] = 1

    return author_list

y_tran = process_label(data_tran)

np.save(PATH_PREFIX + 'data/data_feature/y_tran.npy', y_tran)
```
